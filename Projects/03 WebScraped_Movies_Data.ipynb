```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
headers = {'user-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebkit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.imdb.com/search/title/?title_type=feature&num_votes=25000,&genres=action&sort=user_rating,desc&start=51&ref_=adv_nxt', headers=headers).text
```

```python
soup = BeautifulSoup(webpage, 'lxml')
```

```python
print(soup.prettify())
```

```python
company = soup.find_all('div', class_='lister-item-content')
len(company)
```

```python
Name = []
Year = []
Duretion_min = []
Genre = []
Ratings = []
Votes = []
Income = []
Director = []
Stars = []


for i in company:
  
  Name.append(i.find('a').text.strip())
  Year.append(i.find('span', class_='lister-item-year').text.strip()[1:5])
  Duretion_min.append(int(i.find('span', class_='runtime').text.strip()[0:3]))
  Genre.append(i.find('span', class_='genre').text.strip())
  Ratings.append(float(i.find('strong').text.strip()))
  Votes.append(i.find('p', class_='sort-num_votes-visible').text.split('\n')[2])
  if len(i.find('p', class_='sort-num_votes-visible').text.split('\n'))==6:
    Income.append(i.find('p', class_='sort-num_votes-visible').text.split('\n')[4])
  else:
    Income.append('Nan')
  Director.append(i.find_all('a')[12].text)
  stars = i.find_all('a')[13].text,i.find_all('a')[14].text,i.find_all('a')[15].text, i.find_all('a')[16].text
  Stars.append(stars)


Scraped_Data = {
    'Name':Name,
    'Year':Year,
    'Duretion_min':Duretion_min,
    'Genre':Genre,
    'Ratings':Ratings,
    'Votes':Votes,
    'Income':Income,
    'Director':Director,
    'Stars':Stars
}
```

```python
df = pd.DataFrame(Scraped_Data)
df.shape
```
